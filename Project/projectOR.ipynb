##### Main Refrence of this project: A New Approach to Solve Transportation Problems
##### This refrence is an article by Mollah Mesbahuddin Ahmed, Aminur Rahman Khan, Md. Sharif Uddin, Faruque Ahmed at Department of Mathematics, Jahangirnagar University, Dhaka, Bangladesh

    The transportation problem is a special type of linear programming problem where the objetive consists in minimizing
    transportation cost of a given commodity from a number of sources or origins (e.g. factory, manufacturing facility) to
    a number of destinations (e.g. warehouse, store). Each source has a limited supply (i.e. maximum number of products that
    can be sent from it) while each destination has a demand to be satisfied (i.e. minimum number of products that need to be
    shipped to it). The cost of shipping from a source to a destination is directly proportional to the number of units 
    shipped.
    Finding an initial basic feasible solution is the prime requirement to obtain an optimal solution for
    the transportation problems.

```python
import pandas as pd
import math
df_plant = pd.read_csv("Plant.csv")
df_city = pd.read_csv("City.csv")
```

```python
#Distance Measurement
list_plant_Xcoordinates = df_plant["XCoordinates"].tolist()
list_plant_Ycoordinates = df_plant["YCoordinates"].tolist()
list_plant_Xcoordinates = list(map(float, list_plant_Xcoordinates)) 
list_plant_Ycoordinates = list(map(float, list_plant_Ycoordinates)) 

list_city_Xcoordinates = df_city["XCoordinates"].tolist()
list_city_Ycoordinates = df_city["YCoordinates"].tolist()
list_city_Xcoordinates = list(map(float, list_city_Xcoordinates)) 
list_city_Ycoordinates = list(map(float, list_city_Ycoordinates)) 
```

```python
def distanceCalculator(x1,y1,x2,y2):
    distance = math.sqrt(abs((x1-x2)**2 + (y1-y2)**2))
    return distance
```

```python
distance_allPlantsFromCities = []
for i in range(len(list_plant_Xcoordinates)):
    distance_PlantFromCities = []
    for j in range(len(list_city_Xcoordinates)):
        distance = distanceCalculator(list_plant_Xcoordinates[i], list_plant_Ycoordinates[1],
                                      list_city_Xcoordinates[j], list_city_Ycoordinates[j]);
        distance_PlantFromCities.append(distance) 
    distance_allPlantsFromCities.append(distance_PlantFromCities)
distance_allPlantsFromCities
```

##  <font color='green'>Section D:</font>
### Defining Cost


To determine the cost function, we have to consider the loss of transmitting lines. Obviously, the more the distance, the more the electrical resistance of the line and the more the loss! To obtain a valid cost function we use the distance of transmitting lines and the transmitted power from sources to cities. Cost of using each line will be related to the length of line in a linear way like below:

![title](img/Furmola1.png) 



```python
costs = []
max_dist = max(distance_PlantFromCities)
for i in range(len(distance_allPlantsFromCities)):
    cost_plant_to_city = []
    for j in range(len(distance_allPlantsFromCities[0])):
        cost = ((distance_allPlantsFromCities[i][j])/max_dist)/3
        cost_plant_to_city.append(cost)
    costs.append(cost_plant_to_city)
```
